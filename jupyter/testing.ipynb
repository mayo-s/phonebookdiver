```python
def build_address_hash(address):
    ln, fn, zip, city, st, stn, ac = '', '', '', '', '', '', ''
    if address.get('lastname') is not None: ln = address.get('lastname')
    if address.get('firstname') is not None: fn = address.get('firstname')
    if address.get('zip') is not  None: zip = address.get('zip')
    if address.get('city') is not None: city = address.get('city')
    if address.get('street') is not None: st = address.get('street')
    if address.get('street_number') is not None: stn = address.get('street_number')
    if address.get('area_code') is not None: ac = address.get('area_code')
    if address.get('phonenumber') is not None: pn = address.get('phonenumber')
    address_hash = (ln + fn + zip + city + st + stn + ac + pn).lower().__hash__()
    return address_hash, address
```

```python
import re
def clean_street_info(data_df):
    new_data = []
    for i, d in data_df.iterrows():
        street = d['street']
        if ';' in street:
            street = re.sub('[0-9]+', '', street)
            d['street'] = street[:len(street) - 1]
        new_data.append(d)
    return new_data
```
